## 聚类稳定性

```python
cluster_nmi=pd.read_csv("cluster_nmi.csv")
palette = sns.color_palette("summer",9)#summer pink
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))
g=sns.lineplot(x="epoch", y="NMI",hue="Backbone",#style="Modality",
                errorbar=("sd",2),
                palette=palette, data=cluster_nmi)
                
g.set(title="Cluster Reassginment")
```

## 聚类质量

```python
clusterlabel_nmi=pd.read_csv("clusterlabel_nmi.csv")
palette = sns.color_palette("pink",8)
plt.figure(figsize=(8, 5))
g=sns.lineplot(x="epoch", y="NMI",hue="Backbone",
                errorbar=("sd",1),#markers=True, dashes=False,
                palette=palette,data=clusterlabel_nmi)
g.set(ylim=(0.01,0.18))
g.set(title="Cluster Quality")
```

## K值的影响

```python
k_shot=pd.read_csv("k_shot.csv")
palette = sns.color_palette("mako_r", 6)
sns.set_theme(style="whitegrid")

g=sns.lineplot(x="K",y="5/10-shot",
            errorbar=("sd",0.6),markers=True,style="few-shot",palette=palette,dashes=False,data=k_shot)
plt.xticks([1,2,3,4,5,6],[4,8,16,32,64,128])
g.set(title="Influence of K")
```

```python
g=sns.lineplot(x="K",y="NMI",hue="Modality",style="Modality",
            errorbar=("sd",1),
            markers=True, dashes=False,data=k_nmi)
plt.xticks([1,2,3,4,5,6],[4,8,16,32,64,128])
g.set(title="Cluster Quality of Diffurent K")
```

## 不同模型的散点图

```python
backbone_shot=pd.read_csv("backbone_shot.csv")
sns.set_theme(style="whitegrid")
fig=plt.figure(figsize=(6, 3))
palette = sns.color_palette(palette="Set3", n_colors=10)#"Set3""Pastel2"
g=sns.relplot(x='Params[M]', y="Few-shot", hue="Backbone",
              data=backbone_shot,palette=palette, size='Params[M]',sizes=(200,900))
g.set(ylim=(50,70))
g.set(title="Few-shot Mean Accuracy")
# data=accuracy_top1.drop(3)
# sns.lineplot(x='Params/M', y='Top1_Accuracy', hue="Arch",
#               data = accuracy_top1, palette=["gray"])
# g2 = g.twinx()
# g2 = g.twiny()
```
